```python
from typing import Tuple


def some_functions_1(x: int = 1, y: int = 2) -> Tuple[int, int]:
    """Default settings check."""
    return x, y

print(some_functions_1())


def some_functions_2(x=1, y=2):
    """Default settings check."""
    return x, y


print(some_functions_1())
print(some_functions_1(22, 33), end='\n\n')

print(some_functions_2())
print(some_functions_2(22, 33))


def whats_on_the_telly(penguin=None):
    """
    Default parameter values are evaluated from left to
    right when the function definition is executed.
    This means that the expression is evaluated once,
    when the function is defined, and that the same
    “pre-computed” value is used for each call.
    This is especially important to understand when a default
    parameter is a mutable object, such as a list or
    a dictionary: if the function modifies the object
    (e.g. by appending an item to a list), the default value
    is in effect modified. This is generally not what was
    intended. A way around this is to use None as the default,
    and explicitly test for it in the body of the function, e.g.:
    """
    if penguin is None:
        penguin = []
    penguin.append("property of the zoo")
    return penguin


print(whats_on_the_telly())
```
